```python
import numpy as np
from sympy.plotting import plot
from sympy import *
```

```python
x = sympy.symbols('x')
```

```python
f = -12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
f
```

```python
sympy.solveset(f)
```

Бесконечное множество корней. 

Интервалов возрастания и убывания бесконечное множество.

Бесконечное множество промежутков, где f > 0 и f < 0.

```python
plot(f, (x, -100, 100), line_color = 'red', title = 'График')
```

Определим промежутки, на котором f > 0 и f < 0 в диапазоне (-10, 10) с шагом 1:

```python
x = np.arange(-10, 10, 1)
plus_list = []
minus_list = []

for i in x:
    add = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if add > 0:
        plus_list.append(round(add, 2))
    else:
        minus_list.append(round(add, 2))
print('В диапазоне (-10, 10):\n')
print(f'Промежуток, на котором f > 0: ({plus_list[-1]} -> {plus_list[0]})\n{plus_list}\n')
print(f'Промежуток, на котором f < 0: ({minus_list[0]} -> {minus_list[-1]})\n{minus_list}')
```
